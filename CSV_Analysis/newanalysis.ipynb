```python
import pandas as pd
df = pd.read_csv("https://waf.cs.illinois.edu/discovery/gpa.csv")
df
```

```python
df2 = df[((df.Year == 2023) & (df.Term == 'Spring')) | ((df.Year == 2022) & (df.Term == 'Fall'))]
df2
dfcheck = df2[df2.Subject == 'AAS']
dfcheck
```

```python
df2['Class'] = df2['Subject'] + " " + df2['Number'].astype(str)
df3 = df2.groupby('Class').sum()
df3['GPA'] = (4 * (df3['A+'] + df3['A']) + (3.67 * df3['A-']) 
+ (3.33 * df3['B+']) + (3 * df3['B']) + (2.67 * df3['B-'])
+ (2.33 * df3['C+']) + (2 * df3['C']) + (1.67 * df3['C-'])
+ (1.33 * df3['D+']) + (1 * df3['D']) + (.67 * df3['D-']))/df3['Students']
df3
```

```python
df3.to_csv('gpaData.csv')


#'https://www.reddit.com/r/UIUC/search/?q=' + Class + Number +
#'&type=link&cId=b32ceb77-1b8c-40e5-8b34-d2b4fec8c4b6&iId=bfd40497-f1e5-4941-9509-aa6fccd80b9e'
```
